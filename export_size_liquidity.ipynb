# Export SIZE & LIQUIDITY for a Given Date
This notebook uses cb_cache to read SIZE and LIQUIDITY from daily.hdf.

```python
import os
import sys
import numpy as np
import pandas as pd

sys.path.insert(0, '/home/ubuntu/code/cb_cache')
import cb_cache as cbc
```

```python
# Parameters
trade_date = 20251224
output_path = f'/home/ubuntu/scripts/cache_all/size_liquidity_{trade_date}.csv'
```

```python
# Load symbols via cb_cache and read factors from daily.hdf
symbols = cbc.get_all_symbols()
cache = cbc.EqCache(symbols, trade_date, trade_date, root_path='/home/ubuntu/scripts/cache_all')

size = cache.daily.SIZE[0, :]
liquidity = cache.daily.LIQUIDITY[0, :]

df = pd.DataFrame({
    'date': trade_date,
    'symbol': symbols,
    'SIZE': size,
    'LIQUIDITY': liquidity,
})

df.to_csv(output_path, index=False)
output_path
```
